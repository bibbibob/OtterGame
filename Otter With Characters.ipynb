```python
#Running the game commands
def Initialize(num_players):
    #Shuffle cards up
    char_deck = Shuffle(character_cards)
    spell_deck = Shuffle(spell_cards)
    task_deck = Shuffle(task_cards)
    task_board = []
    discard_pile = []
    Gamers = [Player() for k in range(num_players)]
    #Draw a hand and deck for each player and add them to the gamers list
    for i,Dawg in enumerate(Gamers):
        Char = []
        Draw(1,Char,char_deck)
        Dawg.char = Char[0]
        Draw(7,Dawg.hand,spell_deck)
    Draw(9,task_board,task_deck) #Draw 9 tasks
    print([Dude.char.name for Dude in Gamers])
    return Gamestate(spell_deck,task_deck, task_board, discard_pile, Gamers) #return players and gamestate

def Turn(Game):
    for Player in Game.players:
        Player.char.Main(Player,Game)
        Player.char.Wrap_up(Player,Game)

def Play_Game(NumPlayers, debug=False):
    Game = Initialize(NumPlayers)
    
    for Player in Game.players:
        Player.char.Pregame(Player,Game)
    Turn_Counter = 0
    while Game.end_game == False:
        Turn(Game)
        Turn_Counter += 1
        if Turn_Counter > 100:
            #return Game   #Occationally these games reach some sort of mathemathical stall endpoint I guess. Deal with it.
            Game.end_game = True
    for Player in Game.players:
        try:
            Player.char.Endgame(Player,Game)
        except:
            pass
    if debug==True:
        print("Cards in task deck: ",len(Game.task_deck))
        print([(k.char.name,k.points) for k in Game.players])
        return Game
    return [(k.char.name,k.points) for k in Game.players]
    
```

```python
import io
from contextlib import redirect_stdout

def suppress_print(func):
    """
    A decorator that suppresses print statements for the decorated function.
    """
    def wrapper(*args, **kwargs):
        with io.StringIO() as buf, redirect_stdout(buf):
            return func(*args, **kwargs)
    return wrapper
```

```python
import io
import time as tt
from contextlib import redirect_stdout

t = tt.time()

Stats = {}
for card in character_cards:
    Stats[card.name] = [0, 0]  # Games, #Wins

for k in range(30000):
    if k%1000==0:
        print(k)
    with io.StringIO() as buf, redirect_stdout(buf):
        Results = Play_Game(3)
        # The print statement's output will be suppressed
        print(Results)
    
    i_win = 0
    score = 0
    for Person in Results:
        if Person[1] > score:
            score = Person[1]
            i_win = Results.index(Person)
        Stats[Person[0]][0] += 1
    Stats[Results[i_win][0]][1] += 1
print(tt.time() - t)
```

```python
for Character in list(Stats.keys()):
    print(Character+" winrate: ",np.round(Stats[Character][1]/Stats[Character][0],2),f" across {Stats[Character][0]} games")
```
